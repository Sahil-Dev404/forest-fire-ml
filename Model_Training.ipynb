```python
import pandas as pd 
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline 
```

```python
df=pd.read_csv('Algerian_forest_fires_cleaned_dataset.csv')
```

```python
df.head()
```

```python
df.columns
```

```python
df.drop(['day','month','year'],axis=1,inplace=True)
```

```python
df.head()
```

```python
df['Classes'].value_counts()
```

```python
#Encoding the fire and not fire to 1 and 0
df['Classes']=np.where(df['Classes'].str.contains('not fire'),0,1)
```

```python
df.tail()
```

```python
df['Classes'].value_counts()
```

```python
X=df.drop('FWI',axis=1)
y=df['FWI']
```

```python
X.head()
```

```python
df.head()
```

```python
y #it has the fwi values in the y
```

```python
#Train-Test-Split
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=42)
```

```python
X_train.shape,X_test.shape
```

```python
X_train.corr()
```

```python
plt.figure(figsize=(10,8))
corr=X_train.corr()
sns.heatmap(corr,annot=True)
plt.show()
```

```python
def correlation(dataset , threshold):
    col_corr=set()
    corr_matrix=dataset.corr()
    for i in range(len(corr_matrix.columns)):
        for j in range(i):
            if abs(corr_matrix.iloc[i, j]) > threshold: 
                colname = corr_matrix.columns[i]
                col_corr.add(colname)

    return col_corr      
```

```python
corr_features=correlation(X_train,0.85)
```

```python
#drop features when correlation is more than 0.85
X_train.drop(corr_features,axis=1,inplace=True)
X_test.drop(corr_features,axis=1,inplace=True)
X_train.shape,X_test.shape
```

Standardisation the dataset

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
X_train_scaled=scaler.fit_transform(X_train)
X_test_scaled=scaler.transform(X_test)
```

```python
X_train_scaled
```

```python
X_test_scaled
```

THE EFFECT OF STANDARD SCALER ON THE DATASET

```python
sns.boxplot(data=X_train)
plt.title('X train before scaling')
plt.show()
sns.boxplot(data=X_train_scaled)
plt.title('X tarin after scaling')
plt.show()
```

LINEAR REGRESSION MODEL

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import r2_score
linReg=LinearRegression()
linReg.fit(X_train_scaled,y_train)
y_pred=linReg.predict(X_test_scaled)
mae=mean_absolute_error(y_test,y_pred)
print('mean absolute error:',mae)
score=r2_score(y_test,y_pred)
print('the r2 score:',score)
plt.scatter(y_test,y_pred)
plt.show()
```

LASSO REGRESSION 

```python
from sklearn.linear_model import Lasso
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import r2_score
lasso=Lasso()
lasso.fit(X_train_scaled,y_train)
y_pred=lasso.predict(X_test_scaled)
mae=mean_absolute_error(y_test,y_pred)
score=r2_score(y_test,y_pred)
print("Mean absolute error", mae)
print("R2 Score", score)
plt.scatter(y_test,y_pred)
plt.show()
```

Cross validation for LASSO

```python
from sklearn.linear_model import LassoCV
lassocv=LassoCV(cv=5)
lassocv.fit(X_train_scaled,y_train)
```

```python
lassocv.alpha_
```

```python
lassocv.alphas_
```

```python
y_pred=lassocv.predict(X_test_scaled)
plt.scatter(y_test,y_pred)
plt.show()
mae=mean_absolute_error(y_test,y_pred)
score=r2_score(y_test,y_pred)
print("Mean absolute error", mae)
print("R2 Score", score)
```

->REASON FOR THE CHANGE OF VALUES FROM NORMAL TO CROSS VALIDATION TRAINING
Data is split into k different train/validation sets
Model is retrained k times
Each fold has:
-different samples
-different noise
-different outliers
Scores are averaged

RIDGE REGRESSION MODEL

```python
from sklearn.linear_model import Ridge
from sklearn.metrics import mean_absolute_error, r2_score
ridge=Ridge()
ridge.fit(X_train_scaled,y_train)
y_pred=ridge.predict(X_test_scaled)
mae=mean_absolute_error(y_test,y_pred)
score=r2_score(y_test,y_pred)
print('mean absolute error:',mae)
print('r2 score:',score)
plt.scatter(y_test,y_pred)
plt.show()
```

```python
from sklearn.linear_model import RidgeCV
ridgecv=RidgeCV(cv=5)
ridgecv.fit(X_train_scaled,y_train)
y_pred=ridgecv.predict(X_test_scaled)
plt.scatter(y_test,y_pred)
mae=mean_absolute_error(y_test,y_pred)
score=r2_score(y_test,y_pred)
print("Mean absolute error", mae)
print("R2 Score", score)
plt.show()
```
